## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

```python
url_template = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label={concept}&match={match}&limit={limit}"

concept = "Aspirin"
match = "exact"
limit = "10"

concept = concept.replace(" ", "%20")

url = url_template.replace("{concept}", concept)\
                  .replace("{match}", match)\
                  .replace("{limit}", limit)

data = requests.get(url)
json_mi = data.json()
print("Busca por:"+json.dumps(json_mi[0]["label"], indent=3))


uri_mesh = json_mi[0]["resource"]

bar = str.rindex(uri_mesh, "/")
print("Código da Aspirina: " + uri_mesh[bar+1:])
url = "https://id.nlm.nih.gov/mesh/lookup/details?descriptor={code}"

code = uri_mesh[bar+1:]

url2 = url.replace("{code}", code)

data = requests.get(url2)
json_mid = data.json()
# print(json.dumps(json_mid["terms"], indent=3))


terms=json_mid["terms"]
for i,item in enumerate(terms):
    print("Termo {}: {}".format(i,item["label"]))
```
